## 01A. Creating subset of merged dataset/ Creating new a column using if-statements with user-defined functions (price_range)

```python
# Creates subset of the dataset to only show the first million rows
df = orders_products_merged[:1000000]
```

```python
# Shows row and column count for dataset
df.shape
```

```python
# Assigns products under a certain price range into a low, mid, or high range product category
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

```python
# Applies each price label for all rows under a new column 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Shows counts of values for the created product price ranges
df['price_range'].value_counts(dropna = False)
```

```python
# Shows max price within the 'df' subset
df['prices'].max()
```

## 01B. If-statements with the 'loc()' function

```python
# Assigns products under a certain price range into a low, mid, or high range product category but with loc() function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Shows counts of values for the created product price ranges
df['price_range_loc'].value_counts(dropna = False)
```

## 01C. If-statements with for-loops (busiest_day)

```python
# Shows counts of values for each day of the week
orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []
```

```python
# Assigns days of the week based on count of values (how busy each day is) into least busy, regularly busy, and busiest day categories 
for value in orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Shows list of results
result
```

```python
# Assigns results into a new category of the dataset
orders_products_merged['busiest_day'] = result
```

```python
# Shows counts for each category of busy days
orders_products_merged['busiest_day'].value_counts(dropna = False)
```

## 02. Update 'busiest day' column name and new labels should correspond with two busiest days/two slowest days

```python
new_result = []
```

```python
# Assigns days of the week based on count of values (how busy each day is) into slowest, regular, and busiest days categories 
for value in orders_products_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    new_result.append("Busiest days")
  elif value == 4 or value == 3:
    new_result.append("Slowest days")
  else:
    new_result.append("Regular days")
```

## 04. Identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day”

```python
# Shows counts of values for each hour of a day
orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

### *** The top and bottom rows are in sequences of hours so hours 9-16 and 0-6 can be utilized by using less/more than or equal to in the For-loops commands below. The top hour sequence will be the most orders, the bottom hour sequence will be the fewest orders, and the rest will be average orders.

```python
newest_result = []
```

```python
# Assigns days of the week based on count of values (how busy each day is) into slowest, regular, and busiest days categories 
for value in orders_products_merged["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    newest_result.append("Most orders")
  elif value <=6:
    newest_result.append("Fewest orders")
  else:
    newest_result.append("Average orders")
```

```python
# Assigns results into a new category of the dataset
orders_products_merged["busiest_period_of_day"] = newest_result
```
