### -Missing values found: The first_name column is missing quite a few values (11259) since it doesn't allign with the row count in the shape function. However, nothing can be done for this since we can't assume or input anything in their place. Also, no other columns have missing values so we can utilize 'user_id' or common values in other columns to find a customer.

```python
# Displays only rows that are duplicates in df_cus dataset under a new subset
df_cus_dups = df_cus[df_cus.duplicated()]
```

```python
df_cus_dups
```

### - No duplicates were found

```python
# Checks for mixed types. Will display if each column is mixed or consistent
for col in df_cus.columns.tolist():
 weird = (df_cus[[col]].applymap(type) != df_cus[[col]].iloc[0].apply(type)).any(axis = 1)
 if len (df_cus[weird]) > 0:
        print (col, ' mixed')
 else: print(col, ' consistent')
```

```python
# Changing first_name type to str
df_cus['first_name'] = df_cus['first_name'].astype('str')
```

```python
# Now all columns are consistent
for col in df_cus.columns.tolist():
 weird = (df_cus[[col]].applymap(type) != df_cus[[col]].iloc[0].apply(type)).any(axis = 1)
 if len (df_cus[weird]) > 0:
        print (col, ' mixed')
 else: print(col, ' consistent')
```

### - The first_name column was mixed and then chaned to a string in order for the entire dataset to be consistent
