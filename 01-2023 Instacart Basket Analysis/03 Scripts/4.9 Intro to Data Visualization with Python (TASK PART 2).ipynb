## 02. Create a bar chart, a histogram, a scatterplot, and a line chart, using a sample subset for the line chart

### A. Creating bar charts

```python
# Creates bar chart through matplotlib for 'orders_day_of_week' column by frequency (and ordered by day of week)(with colors)
bar = instacart_data['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['purple', 'red', 'green', 'yellow' , 'blue', 'orange', 'pink'])
```

### B. Histogram

```python
# Replaces any prices over 100 as NaNs
instacart_data.loc[instacart_data['prices'] >100, 'prices'] = np.nan
```

```python
# Now the max of the datast is realistic and won't negatively affect analysis
instacart_data['prices'].max()
```

```python
# Creates histogram for 'prices' column by frequency and now under an object name
hist = instacart_data['prices'].plot.hist(bins = 70)
```

## C. Scatterplot

```python
# Creates scatterplot to visualize the distribution of datapoints in price column
scatter = sns.scatterplot(x = 'prices', y = 'prices',data = instacart_data)
```

## D. Line chart

```python
# Creates a random set of numbers with the 'seed' function with a following number to recreate the exact same set of numbers in the future. Then it uses a 70/30 split(0.7)-- Any rows whose assigned number is less than 0.7 are placed in one sample, while any rows whose assigned number is greater than 0.7 are placed in the other, effectively splitting the dataframe into two dataframes at a 70/30 ratio. Also, the 30 percent sample would statistically resemble the 70 percent sample but contain less rows 
np.random.seed(4)
dev = np.random.rand(len(instacart_data)) <= 0.7
```

```python
# Shows list of true and false values for created dataframe above
dev
```

```python
# Creates 10 random numbers, but in the code above, the '.rand(len(ords_prods_merge)) creates random numbers for the entire dataset
np.random.rand(10)
```

```python
# Creates a list holding true/false values to the test np.random.rand() <= 0.7
dev = np.random.rand(len(instacart_data)) <= 0.7
```

```python
# Stores 70% of the sample in the dataframe called 'big'
big = instacart_data[dev]
```

```python
# Stores 30% of the sample in the dataframe called 'small'
small = instacart_data[~dev]
```

```python
# Shows count of rows in dataset
len(instacart_data)
```

```python
# When adding the 70/30 split, it should equal the same amount of rows as the entire dataset
len(big) + len(small)
```

```python
# Creates a new dataframe with certain columns from the 'small' subset from above 
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Creates line chart with newly created dataframe and assigns the x and y axis values
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

## 03. Create a histogram of the “order_hour_of_day” column

## 07. Explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function

```python
# Shows first 20 rows of these two columns
instacart_data[['age','income']].head(20)
```

### Answer: The highest incomes overall come from ages 40-80 years old ranging from around 300k to 600k and are less frequent. The majority of incomes for ages 40-80 years old range from 25k-300k. The customers who are ages 20-40 years old have the highest incomes for that age group at around 200k to 400k and are less frequent. The majority of incomes for ages 20-40 years old range from 25k-200k. Overall, there is a connection between age and income with older ages making far more than younger ages.
